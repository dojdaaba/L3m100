```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# 2.1
def f(x, mu):
    return mu * x * (1 - x)
```

```python
# 2.2
def plot_f(mu):
    x = np.linspace(0, 1, 100)
    plt.plot(x, f(x, mu), label=mu)
if __name__ == '__main__':
    # 2.2
    for mu in [0.1, 1, 2.5, 4]:
        plot_f(mu)
    plt.legend()
    plt.show()

    
    
```

```python
    # 2.3
    # Si mu est > 4, l'application f sort de l'intervalle [0, 1]
```

```python
# 3.1 + 3.2
def simulation(x0, mu, n, m=0):
    s = []
    for _ in range(n):
        s.append(x0)
        x0 = f(x0, mu)
    return np.array(s[m:])
```

```python
  # 3.3
for x0, mu in [(0.5, 0.1), (0.5, 0.5), (0.5, 1.5), (0.5, 2.5), (0.5, 3.5)]:
    plt.plot(simulation(x0, mu, 20), label=(x0, mu))
    plt.legend()
    plt.show()
```

```python
    # 3.4
    # - Si mu est trop faible, la population converge vers 0, si mu est trop important, la population est erratique, sinon elle se stabilise
    # - La valeur initiale x0 influe sur la rapidité et valeur de convergence de la population
    # - Lorsque mu est proche de 4, la valeur de la population varie fortement d'une généraiton à l'autre sans jamais converger
```

```python
# 4.1
def spider(x0, mu, n):
    sim = simulation(x0, mu, n + 1)
    x = np.repeat(sim[:-1], 2)
    y = np.concatenate([[0], x[2:], sim[-1:]])
    return x, y
```

```python
# 4.2
def plot_spider(x0, mu, n):
    x, y = spider(x0, mu, n)
    plot_f(mu)
    plt.plot(np.arange(0, 2), label="Id")
    plt.plot(x, y, label=(x0, mu))
```

```python
# 4.3
for mu in [1.5, 3.1, 3.5, 3.9]:
        plot_spider(0.7, mu, 20)
        plt.legend()
        plt.show()
```

```python
# 5.1
def bifurcation(x0, mu_list, n, m):
    x = np.repeat(mu_list, n - m)
    y = np.concatenate([simulation(x0, mu, n, m) for mu in mu_list])
    return x, y
```

```python
# 5.2
def plot_bifurcation(x0, num=1000, n=200, m=100):
    mu_list = np.linspace(0, 4, num)
    x, y = bifurcation(0.1, mu_list, 200, 100)
    plt.scatter(x, y, edgecolor=None, s=1)
```

```python
    # 5.3
    plot_bifurcation(0.1)
    plt.show()
```

```python
# 6.1
def attractor(x0, mu, n, m):
    pop = simulation(x0, mu, n + 1, m)
    return pop[:-1], pop[1:]

def plot_attractor(x0, mu, n=5000, m=100):
    x, y = attractor(x0, mu, n, m)
    plt.plot(x, y, '.', label=(x0, mu))
```

```python
# 6.2
for mu in [1.5, 3.1, 3.5, 3.9]:
        plot_attractor(0.7, mu)
        plt.legend()
        plt.show()
```
